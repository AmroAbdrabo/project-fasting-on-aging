```python
def beta_to_m(x):
    res = np.log(x/(1-x))
    if res == -np.inf:
        return np.log((x + 0.0000001)/(1-x + 0.0000001))
    return res
```

```python
buch_cols_m = buch_cols.applymap(beta_to_m)
```

```python
(buch_cols_m == -np.inf).sum().sum()
```

```python
buch_cols_m.to_csv("buch_cols_m.csv")
```
